```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
# Dataframe formating
pd.set_option('display.float_format', lambda x: '%.5f' % x)
plt.rcParams["figure.figsize"] = (20,3)
pd.set_option('display.max_rows', None)
```

```python
hours = {
    '09AM': '-9:00 AM',
    '12AM': '-12:00 PM',
    '15PM': '-15:00 PM',
    '18PM': '-18:00 PM'
}
```

### GFS comparison

```python
GFS_stock_files = sorted(glob('./comparision data/Comparación Febrero/**-Febrero-2021-GFS'))
# removing February 2, since there was no comparison for the NAN and RAP model on this specific date
GFS_stock_files = GFS_stock_files[1:]
GFS_stock_files
```

```python
GFS_feb_03_to_27_comparission = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in GFS_stock_files), ignore_index=True)
GFS_feb_03_to_27_comparission
```

```python
# Get dates for every file in order to add them in the dataframe
dates = [GFS_feb_03_to_27_comparission['filename'][indx].split('/')[-1:][0] for indx in GFS_feb_03_to_27_comparission.index]
```

```python
# Logic in order to add add the hour for every record meassured
for x in range(0, len(dates), 4):
    dates[x] = dates[x] + hours['09AM']
    dates[x + 1] = dates[x + 1] + hours['12AM']
    dates[x + 2] = dates[x + 2] + hours['15PM']
    dates[x + 3] = dates[x + 3] + hours['18PM']
```

```python
# Add dates into dataFrame 
GFS_feb_03_to_27_comparission['Date'] = dates
```

```python
GFS_feb_03_to_27_comparission
```

```python
GFS_feb_03_to_27_comparission.set_index('Date').plot()
plt.ylabel('Power (W)')
plt.title('Comparision Between Pvlib GFS and PV Cutonala from February 03 to 27')
```

```python
GFS_metrics_stock_files = sorted(glob('./comparision data/Comparación Febrero/**-Febrero-2021-GFS-Metrics'))
GFS_metrics_stock_files = GFS_metrics_stock_files[1:]
GFS_metrics_stock_files
```

```python
GFS_metrics_feb_03_to_27 = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in GFS_metrics_stock_files), ignore_index=True)
```
