```python
import pandas as pd
import matplotlib.pyplot as plt
from glob import glob
```

```python
stock_files = sorted(glob('./comparision data/Comparación Febrero/**-Febrero-2021-GFS'))
```

```python
feb_04_to_27_comparission_GFS = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in stock_files), ignore_index=True)
```

```python
feb_04_to_27_comparission_GFS
```

```python
feb_04_to_27_comparission_GFS.plot()
plt.ylabel('Power (W)')
plt.title('Comparision Between Pvlib GFS and PV Cutonala from February 02 to 27')
```
