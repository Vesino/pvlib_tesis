# Diagrama Solar para el centro universitario de Tonalá

### El siguiente ejemplo muestra un uso basico del calculo de la posición solar con pvlib.solarposition.get.solarposition()

```python
from pvlib import solarposition
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
tz = 'America/Mexico_City'
lat, lon = 20.566556, -103.228506
```

```python
times = pd.date_range('2019-01-01 00:00:00', '2020-01-01', closed='left',
                      freq='H', tz=tz)
```

```python
solpos = solarposition.get_solarposition(times, lat, lon)
```

```python
#Remove nightime
solpos = solpos.loc[solpos['apparent_elevation'] > 0, :]
```

```python
ax = plt.subplot(1,1,1, projection='polar')
```

```python
#Dibujar los analemas
points = ax.scatter(np.radians(solpos.azimuth), solpos.apparent_zenith, s=2, label=None, c= solpos.index.dayofyear)
ax.figure.colorbar(points)
```
