```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
study_day = '18'
study_model = 'GFS'
```

```python
comparision_file = './comparision data/Comparación Enero/' + study_day + '-enero-2021-' + study_model
```

```python
metrics_file = './comparision data/Comparación Enero/' + study_day + '-enero-2021-' + study_model + '-Metrics'
```

```python
comparision_data = pd.read_csv(comparision_file)
comparision_data
```

```python
metrics = pd.read_csv(metrics_file)
metrics
```

```python
from glob import glob
```

```python
stock_files = sorted(glob('./comparision data/Comparación Enero/**-Enero-2021-GFS'))
stock_files
```

```python
Jan_18_to_31_comparission_GFS = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in stock_files), ignore_index=True)
```

```python
Jan_18_to_31_comparission_GFS.plot()
plt.ylabel('Power (W)')
plt.title('Comparision Between Pvlib GFS and PV Cutonala from January 18 to 31')
```

```python
metrics_stock_files_GFS = sorted(glob('./comparision data/Comparación Enero/**-Enero-2021-GFS-Metrics'))
metrics_stock_files_GFS
```

```python
Jan_18_to_31_metrics_GFS = pd.concat((pd.read_csv(file).assign(filename = file)
          for file in metrics_stock_files_GFS), ignore_index=True)
```

```python
Jan_18_to_31_metrics_GFS.describe()
```
